```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/emilylin/Downloads/SARPP 2/SASUMMARY__ALL_AREAS_1998_2022.csv')
df.head()
```

```python
#Create a list of all column headers
print('Original dataframe contains columns for:')
columns = list(df.columns)
print(columns)
```

```python
#Get extra detail about original dataset
information = df.info()
information
```

```python
#remove any irrelevant columns and rows that don't contain state or year information
df = df.drop(['GeoFIPS','Region','TableName','LineCode','IndustryClassification','Unit'], axis = 1)
df = df.drop(np.r_[0:15,780:784])
df
```

```python
#keep only wanted columns including state, description, and columns with years 2013-2019
df = df.loc[:, ["GeoName","Description","2013","2014","2015","2016","2017","2018","2019"]]
df
```

```python
#rename column headers to a more recognisable name
df = df.rename(columns = {'GeoName':'State', 'Description':'Variables'})
df
```

```python
#change values in 'Variables' column to a suitable name
replace = ['(millions of chained 2012 dollars) 1/','(millions of constant (2012) dollars) 2/',\
         '(millions of constant (2012) dollars) 3/','4/','5/','6/','7/','8/','9/','10/']

# Function to replace part of the cell with white space
def replace_with_whitespace(cell, replace):
    for string in replace:
        if cell.endswith(string):
            cell = cell.replace(string, ' ' * len(string))
    return cell

# Apply the function to the specified column
df['Variables'] = df['Variables'].apply(replace_with_whitespace, args=(replace,))
df
```

```python
#check for any NULL values that need replacing or removing
print('Number of missing cells is: ')
missing_counts = df.isnull().sum()
print(missing_counts)
```

```python
#check for duplicates in data frame
duplicate = df[df.duplicated()]
print(len(duplicate)) 
```

```python
# Melt the DataFrame to convert years (2013-2019) to a single column
melted_df = pd.melt(df, id_vars=['State', 'Variables'], var_name='Year', value_name='Value')

# Pivot the melted DataFrame to have 'Variables' as columns and 'Year' as a separate column
pivot_df = melted_df.pivot(index=['State','Year'], columns='Variables', values='Value')

# Reset the index to make 'Year' a separate column
pivot_df.reset_index(inplace=True)

# Rename the index column to 'Years'
pivot_df.rename(columns={'Year': 'Year'}, inplace=True)
pivot_df
```

```python
# Rename the first column header
pivot_df = pivot_df.rename_axis(columns={'Variables': ''})
pivot_df
```

```python
#Check datatype of each column
print(pivot_df.dtypes)
```

```python
#Change datatype of columns to integer and floats where necessary 
pivot_df.columns = pivot_df.columns.str.strip()
# using dictionary to convert specific columns
convert_dict = {'Year': int,
                'Disposable personal income': float,
                'Gross domestic product (GDP)': float,
                'Implicit regional price deflator': float,
                'Per capita disposable personal income': int,
                'Per capita personal consumption expenditures (PCE)': int,
                'Per capita personal income': float,
                'Personal consumption expenditures': float,
                'Personal income': float, 
                'Real GDP': float, 
                'Real PCE': float, 
                'Real per capita PCE': int, 
                'Real per capita personal income': int, 
                'Real personal income': float,
                'Regional price parities (RPPs)': float,
                'Total employment (number of jobs)': int,
                }
 
pivot_df = pivot_df.astype(convert_dict)
print(pivot_df.dtypes)
```

```python
#exporting file as an excel
pivot_df.to_excel('/Users/emilylin/Downloads/Data1002_US_income_dataset.xlsx')
print('DataFrame is written to Excel File successfully.')
```

```python
#We aim at investigating state statistics
#Detail emoployment distribution in different counties under the states does not contribute towards our project
#Therefore, we only save the rows with state names contained in the list
#Unnecessary rows describing the county are dropped

state_full=['Alabama','Alaska','American Samoa','Arizona','Arkansas','California','Colorado','Connecticut','Delaware','District of Columbia','Federated States of Micronesia','Florida','Georgia','Guam','Hawaii','Idaho','Illinois','Indiana','Iowa','Kansas','Kentucky','Louisiana','Maine','Marshall Islands','Maryland','Massachusetts','Michigan','Minnesota','Mississippi','Missouri','Montana','Nebraska','Nevada','New Hampshire','New Jersey','New Mexico','New York','North Carolina','North Dakota','Northern Mariana Islands','Ohio','Oklahoma','Oregon','Palau','Pennsylvania','Rhode Island','South Carolina','South Dakota','Tennessee','Texas','Utah','Vermont','Virgin Island','Virginia','Washington','West Virginia','Wisconsin','Wyoming']
df = df[df['Area_Name'].isin(state_full)]
df
```

```python
#Check if there are duplicate rows and drop them if there are any
df=df.drop_duplicates()
df
```

```python
#Check if there are missing values
print("Missing values distribution: ")
print(df.isnull().mean())
print("")
```

```python
#Change the column names to avoid any confusion
df.columns.values[0] = "Postal Abbr."
df.columns.values[1] = "State"
```

```python
#Now we need to convert the dataset from wide to long format so that it is easier to read
#It also provides convenience to the data integration process later on.
#The first step is to reshape the DataFrame with Pandas stack() and unstack()
df1=df[df.columns].set_index(['Postal Abbr.','State']).stack().reset_index()
df1
```

```python
#Rename relevent columns
df1=df1.rename(columns={'level_2': 'Type',0: 'Rate'} )
```

```python
#Seperate the year from the 'Type' column using the rsplit function targeting at the last underscore
#Create a new column called 'Year' to store the data
df1[["Type","Year"]]=df1["Type"].str.rsplit("_",n=1,expand=True)
df1
```

```python
#Move the columns
df1 = df1[['Postal Abbr.','State','Year','Type','Rate']]
df1['Year']=df1['Year'].astype('int')
```

```python
#Create a pivot table using 'Postal Abbr.','State','Year' as index
df1=df1.pivot_table(index=['Postal Abbr.','State','Year'], columns='Type', values='Rate')
df1
```

```python
#Remove the unnecessary heading
df1.columns.name = ' ' 
df1
```

```python
#Pivot table to normal table, sorted by year.
df1=df1.sort_values(by=['State','Year'])
df1 = df1.reset_index()
df1
```

```python
#The reshaping dataframe process is completed
#Check data type again
#Check if the data types are correct
print("Column datatypes: ")
print(df1.dtypes)
```

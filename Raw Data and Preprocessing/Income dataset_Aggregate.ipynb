```python
import pandas as pd 
df = pd.read_excel('/Users/emilylin/Downloads/SARPP 2/SASUMMARY__ALL_AREAS_1998_2022.xlsx')
df
```

```python
df = df.drop(columns = ['Unnamed: 0'])
df
```

```python
# Strip whitespace from column headers
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
# Group by year and calculate the sum of GDP for each year
gdp_sum_by_year = df.groupby('Year')['Real GDP'].sum().reset_index()

# Print or work with the resulting DataFrame
print(gdp_sum_by_year)
```

```python
# Group by year and calculate the mean of the Implicit regional price deflator for each year
IRPD_mean = df.groupby('Year')['Implicit regional price deflator'].mean().reset_index()
print(IRPD_mean)
```

```python
# Compare mean with the median
IRPD_median = df.groupby('Year')['Implicit regional price deflator'].median().reset_index()
print(IRPD_median)
```

```python
df_copy = df.copy()
# Over the years 2013 to 2019, which states appear to have on average the highest real per capita personal income?
df1 = df_copy[["State","Real per capita personal income"]].groupby("State").mean().reset_index()
print("Average real per capita personal income across all US states from 2013-2019 in order:")

# Sort the average per capita PI in descending order
R_PI = df1.sort_values(by='Real per capita personal income', ascending=False)
R_PI
```

```python
# Find the maximum and minimum values in the "Real per capita personal income" column
max_value = R_PI["Real per capita personal income"].max()
min_value = R_PI["Real per capita personal income"].min()

# Find the corresponding names for the maximum and minimum values
max_name = R_PI.loc[R_PI["Real per capita personal income"] == max_value, "State"].values[0]
min_name = R_PI.loc[R_PI["Real per capita personal income"] == min_value, "State"].values[0]

print(f"Maximum Value: {max_value} (State: {max_name})")
print(f"Minimum Value: {min_value} (State: {min_name})")
```

```python
df_copy2 = df.copy()
# Over the years 2013 to 2019, which states appear to have on average the highest real per capita PCE
df2 = df_copy2[["State","Real per capita PCE"]].groupby("State").mean().reset_index()
print("Average real per capita PCE across all US states from 2013-2019 in order:")

# Sort the average per capita PI in descending order
R_PCE = df2.sort_values(by='Real per capita PCE', ascending=False)
R_PCE
```

```python
# Find the maximum and minimum values in the "Real per capita PCE column
max_value = R_PCE["Real per capita PCE"].max()
min_value = R_PCE["Real per capita PCE"].min()

# Find the corresponding names for the maximum and minimum values
max_name = R_PCE.loc[R_PCE["Real per capita PCE"] == max_value, "State"].values[0]
min_name = R_PCE.loc[R_PCE["Real per capita PCE"] == min_value, "State"].values[0]
print("States with the maximum and minimum average value for 'Real per capita PCE'")
print(f"Maximum Value: {max_value} (State: {max_name})")
print(f"Minimum Value: {min_value} (State: {min_name})")
```
